### **Time-series seasonal decomposition**

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse

plt.rcParams.update({'figure.figsize': (10,10)})
y = df_vwap['VWAP'].to_frame()


# Multiplicative Decomposition
result_mul = seasonal_decompose(y, model='multiplicative',period = 52)

# Additive Decomposition
result_add = seasonal_decompose(y, model='additive',period = 52)

# Plot
plt.rcParams.update({'figure.figsize': (5,5)})
result_mul.plot().suptitle('Multiplicative Decompose', fontsize=11)
result_add.plot().suptitle('Additive Decompose', fontsize=11)
plt.show()
```

```python
## Extract the Components
# Actual Values = Product of (Seasonal * Trend * Resid)
df_reconstructed = pd.concat([result_add.seasonal, result_add.trend, result_add.resid, result_add.observed], axis=1)
df_reconstructed.columns = ['seas', 'trend', 'resid', 'actual_values']
df_reconstructed.tail()
```

#**Feature Extraction**

```python
df_vwap = df_vwap.reset_index()
df_vwap['Date'] = pd.to_datetime(df_vwap['Date'],
                                  infer_datetime_format=True)
df_vwap['year'] = df_vwap.Date.dt.year
df_vwap['month'] = df_vwap.Date.dt.month
df_vwap['day'] = df_vwap.Date.dt.day
df_vwap['day of week'] = df_vwap.Date.dt.dayofweek

#Set Date column as the index column.
df_vwap.set_index('Date', inplace=True)
df_vwap.head()
```

#**Rolling windows**


```python
df_vwap['VWAP'].plot(figsize = (10,6))
plt.grid(True)
plt.show()
```

```python
df_vwap_non_numeric = df_vwap.apply(pd.to_numeric, errors='coerce')
df_vwap_non_numeric.rolling(2).mean().head(8)
```

```python
df_vwap_numeric = df_vwap.apply(pd.to_numeric, errors='coerce')
df_vwap_numeric.dropna(inplace=True)
df_vwap['VWAP'].plot()
df_vwap_numeric.rolling(window=30).mean()['VWAP'].plot(figsize=(10, 6))
plt.grid(True)
plt.show()
```
